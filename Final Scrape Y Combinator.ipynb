## Scraped Y Combinator's Top Companies and Added Data Into Excel

```python
import csv
import requests
import urllib.request
from bs4 import BeautifulSoup

url = 'https://www.ycombinator.com/topcompanies/'
response = requests.get(url)
html = response.content

soup = BeautifulSoup(html)
```

```python
rows = soup.findAll("div", class_="detailsBox")
```

```python
file = open('ycombo.csv', 'w')
writer = csv.writer(file)

#write writer row
writer.writerow(['ranks','company', 'founders','sector','jobs_created','company_description','batch', 'link'])

for row in rows:
    rank = row.find('span').text.strip()
    company = row.find('b').text.strip()
    founders = row.p.find_next_sibling().find_next_sibling().text.strip()
    sector = row.findAll('p')[3].text.strip()
    jobs_created = row.findAll('p')[4].text
    company_description = row.findAll('p')[0].text
    batch = row.findAll('p')[7].text
    link = row.find('a').attrs['href']
    
#     print(rank, ' ', company, ' ', founders, ' ', sector, ' ', jobs_created, ' ', company_description
#          ,' ', link)
    writer.writerow([rank, company, founders, sector, jobs_created, company_description, batch, link])
```
